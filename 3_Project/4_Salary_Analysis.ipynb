```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_IL1 = df[df['job_country'] == 'Israel']
df_IL1
```

```python
df_IL = df[df['job_country'] == 'Israel'].dropna(subset=['salary_year_avg'])
df_IL
```

Out of the 4887 job listings in Israel, only 103 showcased the average yearly salary.

```python
df_IL['job_title_short'].value_counts()
```

```python
job_titles_short = df_IL['job_title_short'].value_counts().index[:6].tolist()
df_IL_top6 = df_IL[df_IL['job_title_short'].isin(job_titles_short)]
df_IL_top6

job_order = df_IL_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
sns.boxplot(data=df_IL_top6, x='salary_year_avg', y='job_title_short', order=job_order)
sns.set_theme(style='ticks')
sns.despine()

# this is all the same
plt.title('Salary Distributions of Data Jobs in Israel')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 600000) 
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.show()
```
